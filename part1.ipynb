```python
def countBOW( nlp,  rawtext ):
    rawtext = re.sub( '\s+', ' ', rawtext )
    text = nlp(rawtext) # we use spacy to process the raw text
    bow = Counter( [ token.lemma_ for token in text] )
    return bow
reuters_bow = dict() # stores the bow of each document
DF = Counter() # stores the document freq (DF) of each word

# for each file id in the reuters dataset
for fileid in data['Article Title']:
    # we count bow for the article with that file id, and store its bow model in reuters_bow
    reuters_bow[fileid] = countBOW( nlp, fileid )
    # for each word (lemma) in the current article, we update its DF (document frequency)
    for word in reuters_bow[fileid]:
        DF[word] += 1
print(len(reuters_bow))
print(DF['Unlost'])
N = len(data)
IDF = Counter( { word:math.log(N/DF[word]) for word in DF } )
print(IDF.most_common(50))
print("len of document ", N)
```

```python
def countBOW(  rawtext ):
    rawtext = re.sub( '\s+', ' ', rawtext )
    text = nlp(rawtext) # we use spacy to process the raw text
    bow = Counter( [ token.lemma_.lower() for token in text if not token.is_stop and not token.is_punct ] )
    return bow
reuters_bow = dict() # stores the bow of each document
DF = Counter() # stores the document freq (DF) of each word

# for each file id in the reuters dataset
for fileid in data['Article Title']:
    # we count bow for the article with that file id, and store its bow model in reuters_bow
    reuters_bow[fileid] = countBOW(  fileid )
    # for each word (lemma) in the current article, we update its DF (document frequency)
    for word in reuters_bow[fileid]:
        DF[word] += 1

N = len(data)
print(data['Article Title'].isnull().values.any())
print(N)
IDF = Counter( { word:math.log(N/DF[word]) for word in DF } )       

print(len(reuters_bow))
reuters_bow_tfidf = dict()
for fileid in reuters_bow:
    reuters_bow_tfidf[fileid] = Counter( { word:(reuters_bow[fileid][word]*IDF[word]) for word in reuters_bow[fileid] } )

len(reuters_bow_tfidf)
def cosine( d1, d2 ):
    voc = set(d1.keys()).union(d2.keys())
    top = sum( [ d1[w]*d2[w] for w in voc ] )
    bot1 = sum( [ d1[w]**2 for w in voc ] )
    bot2 = sum( [ d2[w]**2 for w in voc ] )
    if bot1==0 or bot2==0:
        return 0
    return top/((bot1**0.5)*(bot2**0.5))
for fileid in reuters_bow_tfidf:
    d1 = fileid
    break
reuters_cosine = Counter( { fileid:cosine( reuters_bow_tfidf[d1], reuters_bow_tfidf[fileid] ) for fileid in reuters_bow_tfidf if fileid != d1 } )
reuters_cosine.most_common(10)
```
