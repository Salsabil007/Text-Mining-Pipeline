```python
import spacy

nlp = spacy.load( "en_core_web_lg", disable=["parser", "ner"] )
```

```python
import numpy as np
```

```python
import pandas as pd

aa = pd.read_csv('aa.csv')
asian = pd.read_csv('asian.csv')
hispanic = pd.read_csv('hispanic.csv')
white = pd.read_csv('white.csv')
print(white.dtypes)
```

```python
#step1
def percent_50(data):
    #print(data.head(10))
    aa_percent = data
    aa_percent['PercentRace'] = aa_percent['PercentRace'].str.strip('%')
    aa_percent['PercentRace'] = aa_percent['PercentRace'].astype(float)
    #print(aa_percent.dtypes)
    #print(aa_percent.head(20))
    aa_percent = aa_percent[aa_percent.PercentRace >= 50]
    #print(aa_percent.head(10))
    aa_percent['LastName'] = aa_percent['LastName'].str.title()
    return aa_percent
    #print(aa_percent.head(10))
```

```python
#step1
aa = percent_50(aa)
asian = percent_50(asian)
hispanic = percent_50(hispanic)
white = percent_50(white)
#print(white.head(20))
```

```python
#step2
def remove_zero(aa):
    vec = []
    for ind in aa.index:
        st = aa['LastName'][ind]
        if np.all(nlp.vocab[st].vector == 0):
            vec.append("False")
        else:
            vec.append("True")
    aa['vec'] = vec
    aa = aa[aa.vec == "True"]
    aa = aa.drop(['vec'], axis = 1)
    return aa
```

```python
#step2
print(len(aa))
aa = remove_zero(aa)
print(len(aa))

print(len(asian))
asian = remove_zero(asian)
print(len(asian))

print(len(hispanic))
hispanic = remove_zero(hispanic)
print(len(hispanic))

print(len(white))
white = remove_zero(white)
print(len(white))
```

```python
#step 3
def similarity(aa):
    title = ['Dr.', 'Prof.', 'Ms.', 'Mr.', 'dean','chancellor','provost','congrassman','senator','lieutenant','sergeant']
    for i in title:
        print([(token, nlp(token).similarity(nlp(i))) for token in aa['LastName']])
similarity(aa)
similarity(asian)
similarity(hispanic)
similarity(white)
```

```python
d1 = nlp('Jackson')
d2 = nlp('Mr.')
d1.similarity(d2)
```

```python
#step 4
from statistics import mean
title = ['Dr.', 'Prof.', 'Ms.', 'Mr.', 'dean','chancellor','provost','congressman','senator','lieutenant','sergeant']
t = []
afa = []
asn = []
his = []
whi = []
for i in title:
    t.append(i)
    temp = [nlp(token).similarity(nlp(i)) for token in aa['LastName']]
    afa.append(sum(temp)/len(temp))
    temp = [nlp(token).similarity(nlp(i)) for token in asian['LastName']]
    asn.append(sum(temp)/len(temp))
    temp = [nlp(token).similarity(nlp(i)) for token in hispanic['LastName']]
    his.append(sum(temp)/len(temp))
    temp = [nlp(token).similarity(nlp(i)) for token in white['LastName']]
    whi.append(sum(temp)/len(temp))
d = pd.DataFrame()
d["title"] = t
d["African American"] = afa
d["Asian"] = asn
d['Hispanic'] = his
d["White"] = whi
print(d)
```

#A1.1
Even though spacy's word embeddings do not seem like disfavoring any race, but it is quite clear that it has positive racial bias towards White. White has consistently higher similarity with dean, chancellor, congressman, senator, leiutenant, sergeant and Dr compared to other races. Also, the mebedding seems slightly biased against the Asian and Hispanic, though it is not that much prominent and might be ruled out if we check other such similarities.
